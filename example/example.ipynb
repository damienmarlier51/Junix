```python
import matplotlib.pyplot as plt
```

```python
for i in range(0, 5):
    fig, ax = plt.subplots()
    ax.plot([0, i + 1], [0, i + 1])
    plt.show()
```

```python
for i in range(0, 5):
    fig, ax = plt.subplots()
    ax.plot([0, - (i + 1)], [0, i + 1])
    plt.show()
```
