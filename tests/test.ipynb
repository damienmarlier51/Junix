```python
import matplotlib.pyplot as plt
from IPython.display import Image
from IPython.core.display import SVG
```

```python
for i in range(0, 2):
    fig, ax = plt.subplots()
    ax.plot([0, i + 1], [0, i + 1])
    plt.show()
```

```python
for i in range(0, 2):
    fig, ax = plt.subplots()
    ax.plot([0, - (i + 1)], [0, i + 1])
    plt.show()
```
